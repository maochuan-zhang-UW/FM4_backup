```python
from icalendar import Calendar, Event
from datetime import datetime
import pytz

# Create a calendar
cal = Calendar()
cal.add('prodid', '-//Maochuan//NONSGML v1.0//EN')
cal.add('version', '2.0')

# Create an event
event = Event()
event.add('summary', 'General Exam - Maochuan')
event.add('description', 'General exam for Maochuan, including discussion of research proposal. Room to be confirmed.')
event.add('location', 'TBD (In-Person, Room to be Confirmed)')
event.add('dtstart', datetime(2025, 8, 15, 10, 0, 0, tzinfo=pytz.timezone('America/Los_Angeles')))
event.add('dtend', datetime(2025, 8, 15, 12, 0, 0, tzinfo=pytz.timezone('America/Los_Angeles')))
event.add('dtstamp', datetime(2025, 5, 20, 17, 30, 0, tzinfo=pytz.UTC))
event.add('uid', '20250520T173000Z-123401@maochuan-exam')
event.add('organizer', 'mailto:maochuan@example.com')

# Add the event to the calendar
cal.add_component(event)

# Save the calendar to a file
with open('general_exam.ics', 'wb') as f:
    f.write(cal.to_ical())
```
